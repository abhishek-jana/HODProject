```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import os
```

```python
wp_path = '/mnt/data4/Abhishek/fidmock/wp'
vpf_path = '/mnt/data4/Abhishek/fidmock/vpf'
cic_path = '/mnt/data4/Abhishek/fidmock/cic'
cia_path = '/mnt/data4/Abhishek/fidmock/cia'
```

```python
wp_filenames = ['wp_subsample_'+f'{files:04d}'+'.npy' for files in range(100)]
wp = []
for files in wp_filenames:
    sample = np.load(os.path.join(wp_path,files))
    wp.append(sample[0])
wp_z = np.vstack(wp).T  
wp_filenames = ['wp_subsample_'+f'{files:04d}'+'.npy' for files in range(100,200)]
wp = []
for files in wp_filenames:
    sample = np.load(os.path.join(wp_path,files))
    wp.append(sample[0])
wp_x = np.vstack(wp).T 
wp_filenames = ['wp_subsample_'+f'{files:04d}'+'.npy' for files in range(200,300)]
wp = []
for files in wp_filenames:
    sample = np.load(os.path.join(wp_path,files))
    wp.append(sample[0])
wp_y = np.vstack(wp).T 
```

```python
wp = (wp_x + wp_y + wp_z)/3
```

```python
vpf_filenames = ['vpf_subsample_'+f'{files:04d}'+'.npy' for files in range(100)]
vpf = []
for files in vpf_filenames:
    sample = np.load(os.path.join(vpf_path,files))
    vpf.append(sample[1])
vpf_z = np.vstack(vpf).T  
vpf_filenames = ['vpf_subsample_'+f'{files:04d}'+'.npy' for files in range(100,200)]
vpf = []
for files in vpf_filenames:
    sample = np.load(os.path.join(vpf_path,files))
    vpf.append(sample[1])
vpf_x = np.vstack(vpf).T 
vpf_filenames = ['vpf_subsample_'+f'{files:04d}'+'.npy' for files in range(200,300)]
vpf = []
for files in vpf_filenames:
    sample = np.load(os.path.join(vpf_path,files))
    vpf.append(sample[1])
vpf_y = np.vstack(vpf).T 
```

```python
vpf = (vpf_x + vpf_y + vpf_z)/3
```

```python
cic_filenames = ['cic_subsample_'+f'{files:04d}'+'.npy' for files in range(100)]
cic = []
for files in cic_filenames:
    sample = np.load(os.path.join(cic_path,files))
    hist_cic,_ = np.histogram(sample,bins=30)
    hist_cic = hist_cic/np.sum(hist_cic)
    del sample
    cic.append(np.log(hist_cic))
cic_z = np.vstack(cic).T 
cic_filenames = ['cic_subsample_'+f'{files:04d}'+'.npy' for files in range(100,200)]
cic = []
for files in cic_filenames:
    sample = np.load(os.path.join(cic_path,files))
    hist_cic,_ = np.histogram(sample,bins=30)
    hist_cic = hist_cic/np.sum(hist_cic)
    del sample
    cic.append(np.log(hist_cic))
cic_x = np.vstack(cic).T
cic_filenames = ['cic_subsample_'+f'{files:04d}'+'.npy' for files in range(200,300)]
cic = []
for files in cic_filenames:
    sample = np.load(os.path.join(cic_path,files))
    hist_cic,_ = np.histogram(sample,bins=30)
    hist_cic = hist_cic/np.sum(hist_cic)
    del sample
    cic.append(np.log(hist_cic))
cic_y = np.vstack(cic).T
```

```python
cic = (cic_x + cic_y + cic_z)/3
```

```python
cia_filenames = ['cia_subsample_'+f'{files:04d}'+'.npy' for files in range(100)]
cia = []
for files in cia_filenames:
    sample = np.load(os.path.join(cia_path,files))
    hist_cia,_ = np.histogram(sample,bins=30)
    hist_cia = hist_cia/np.sum(hist_cia)
    del sample
    cia.append(np.log(hist_cia))
cia_z = np.vstack(cia).T 
cia_filenames = ['cia_subsample_'+f'{files:04d}'+'.npy' for files in range(100,200)]
cia = []
for files in cia_filenames:
    sample = np.load(os.path.join(cia_path,files))
    hist_cia,_ = np.histogram(sample,bins=30)
    hist_cia = hist_cia/np.sum(hist_cia)
    del sample
    cia.append(np.log(hist_cia))
cia_x = np.vstack(cia).T 
cia_filenames = ['cia_subsample_'+f'{files:04d}'+'.npy' for files in range(200,300)]
cia = []
for files in cia_filenames:
    sample = np.load(os.path.join(cia_path,files))
    hist_cia,_ = np.histogram(sample,bins=30)
    hist_cia = hist_cia/np.sum(hist_cia)
    del sample
    cia.append(np.log(hist_cia))
cia_y = np.vstack(cia).T 
```

```python
wpz_cov = np.cov(wp_z, bias = True)
wpx_cov = np.cov(wp_x,bias = True)
wpy_cov = np.cov(wp_y,bias = True)
wp_cov = (wpz_cov + wpx_cov + wpy_cov)/3
```

```python
wpz_corr = np.corrcoef(wp_z)
wpx_corr = np.corrcoef(wp_x)
wpy_corr = np.corrcoef(wp_y)
wp_corr = (wpz_corr + wpx_corr + wpy_corr)/3
```
